```python
def min_fun(part):#поиск минимального объекта
    cur=1000000
    for one in part:
        if one>=0:
            if one < cur:
                cur = one
    return cur
```

```python
def max_fun(part):#поиск максимального объекта
    cur=-1000000
    for one in part:
        if one<=0:
            if one > cur:
                cur = one
    return cur
```

```python
def first(maxx, minn, med, part):#всего есть 6 различных вариантов при смене крайних элементов
    part.pop(0)                  #первый из них - когда число которое было слева было меньше медианы,
    minn = med                   #а второе больше, тогда медианой становится число следующее по возрастанию после 
    med = maxx                   #по возрастанию после медианы, число чуть меньше медианы станет самим значением 
    part.remove(med)             #предыдущей медианы и т.д.
    part = np.array(part)
    part = part - med
    maxx = min_fun(part)+med
    return maxx, minn, med
```

```python
def second(maxx, minn, med, part):
    part.remove(med)
    part = np.array(part)
    part = part - med
    maxx = min_fun(part[1:])+med
    return maxx, minn, med
```

```python
def second_half(maxx, minn, med, part):
    part.remove(med)
    part = np.array(part)
    part = part - med
    minn = max_fun(part[1:])+med
    return maxx, minn, med
```

```python
def third(maxx, minn, med, part):

    maxx = med
    med = minn
    part.remove(med)
    part = np.array(part)
    part = part - med
    minn = max_fun(part[1:])+med
    return maxx, minn, med
```

```python
def fourth(maxx, minn, med, part):
    med = minn
    part.remove(med)
    part = np.array(part)
    part = part - med
    minn = max_fun(part[1:])+med
    return maxx, minn, med
```

```python
def fifth(maxx, minn, med, part):
    med = maxx
    part.remove(med)
    part = np.array(part)
    part = part - med
    maxx = min_fun(part[1:])+med
    return maxx, minn, med
```

```python
def mediana(maxx, minn, med, part): 
    part1 = list()
    for a in part:                 #здесь я распределяю события по функциям и учитываю, что число, которое
        part1.append(a)            #только попало в окно сразу может быть ближайшим слева или справа, а также
    if part[0] == maxx:            #что число которое вышло за пределы окна могло быть minn or maxx
        part1.remove(med)
        part1 = np.array(part1)
        part1 = part1 - med
        maxx = min_fun(part1)+med
    
    if part[0] == minn:
        part1.remove(med)
        part1 = np.array(part1)
        part1 = part1 - med
        minn = max_fun(part1)+med
    
    if part[-1:][0] > minn and part[-1:][0]< med:
        minn = part[-1:][0]
    if part[-1:][0] > med and part[-1:][0]< maxx:
        maxx = part[-1:][0]
    
    if med<part[0]:
        if med>part[-1:][0]:
            maxx, minn, med = third(maxx, minn, med, part)
            
        elif med<part[-1:][0]:
            maxx, minn, med = second(maxx, minn, med, part)
            
    elif med>part[0]:
        if med>part[-1:][0]:
            maxx, minn, med = second_half(maxx, minn, med, part)
            
        elif med<part[-1:][0]:
            maxx, minn, med = first(maxx, minn, med, part)
            
    else:
        if med>part[-1:][0]:
            maxx, minn, med = fourth(maxx, minn, med, part)
                        
        else:
            maxx, minn, med = fifth(maxx, minn, med, part)
    return maxx, minn, med
```
